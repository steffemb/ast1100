```python
def readfile(star):
    infile = open(star, 'r')

    times = []
    lamdas = []
    fluxs = []

    while True:
        line = infile.readline()
        if not line:
            break
        numbers = line.split()
        time = float(numbers[0])
        lamda = float(numbers[1])
        flux = float(numbers[2])
    
        times.append(time)
        lamdas.append(lamda)
        fluxs.append(flux)
    infile.close()
    
    return [times, lamdas, fluxs]
    
#x = linspace(0, len(fluxs), len(fluxs))

#plot(x, lamdas)
```

```python
C = 299792458 # lightspeed
lamda0 = 656.3*10**(-9) #m
G = 6.67384*10**(-11) 
Ms = 0.8*1.9891*10**30 # solar masses
```

```python
def Vr(lamda):
    Vr = []
    i = 0
    for i in range(0, len(lamda)):
        Vr.append( float((lamda[i] - lamda0)/lamda0)*C )
        i = i + 1
    return Vr
```

```python
star = 'star0.txt'

data = readfile(star)

VrStar = Vr(data[1])
VrStarP = sum(VrStar)/len(VrStar)
VrStarR = (VrStar - VrStarP) # relative
t = data[0]
    
plot(t, VrStarR)

#print VrStarMean
```
