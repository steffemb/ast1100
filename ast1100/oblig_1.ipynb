I chose to use lists and the ".append" command instead of vecrorizing with linspace. I did this because it's no way to know how many steps the program needed to take.
The vectors i need is $\vec{rx}$, $\vec{ry}$, and $\vec{vx}$, $\vec{vy}$ for both the lander and mars. I define the vectors as list where the $\vec{r}$ vectors give me the length from origo, and the $\vec{v}$ vectors give me the magnitude and direction of $\vec{Vcm}$.

```python
altitude = 298000 # initial altitude
velocity = 4000 # beagle 2
rm = 3400000 #radius mars
mm = 6.4*10**23 # mass mars
ms = 100 # mass satelite (beagle 2)
G = 6.67384*10**(-11)

dt = 1

r2x = [-(altitude + rm)] # m
r2y= [0] # m
v2x = [0] # m/s
v2y = [-velocity] # m/s
r1x = [0] # m
r1y = [0] # m
v1x = [0] # m/s
v1y = [0] # m/s
```

In this block i only define the forces as functions. I chose to decompose the forces in the x-y plane and define one function for each force. The tricky part is to solve $\vec{F}$ = $\frac{G*Mm*Ms}{r**2}$ $\hat{r}$ = $\frac{G*Mm*Ms}{r**2}$ $\frac{\vec{r}}{r}$ = $\frac{G*Mm*Ms}{r**3} (rx\hat{i} + ry\hat{j})$

The force of friction can be writen straight out of the formula without anny "if-test". This is because I have vectorized everything. If i get a negative value of $\vec{v}$ this means the object is moving in "negative" direction, when input in the formula for friction i get back a positive value, and all is fine.

```python
def Fmx(G, mm, ms, rx):
    Fmx = ((G*mm*ms)/r**3)*rx
    return Fmx

def Fmy(G, mm, ms, ry):
    Fmy = ((G*mm*ms)/r**3)*ry
    return Fmy

def Fsx(G, mm, ms, rx):
    Fsx = -((G*mm*ms)/r**3)*rx
    return Fsx

def Fsy(G, mm, ms, ry):
    Fsy = -((G*mm*ms)/r**3)*ry
    return Fsy

def fx(vx2):
    k = 0.00016 #kg/s
    fx = -k*vx2
    return fx

def fy(vy2):
    k = 0.00016 #kg/s
    fy = -k*vy2
    return fy
```

This is the Euler-Chromer block. First i define rx, ry = $|\vec{rx}|, |\vec{ry}|$, to be the distance between the center of mass of the two objects in the x, and y direction. And r to be $|\vec{r}|$ the total length between them.

In the Euler-Chromer i used Newtons 2. law to find $\vec{a}$

$\sum{\vec{F}} = m\vec{a}$

wehere $\sum{\vec{F}}$ = $\vec{Fg}+\vec{Ff}$ , for the beagle 2.

```python
r = altitude + rm

i = 0
while r > rm :
    
    rx = r2x[i] - r1x[i]
    ry = r2y[i] - r1y[i]
    r = sqrt(rx**2 + ry**2)
    
    a2x = (Fsx(G, mm, ms, rx) + fx(v2x[i]))/ms
    vx2 = v2x[i] + a2x*dt
    x2 = r2x[i] + vx2*dt
    r2x.append(x2) # m
    v2x.append(vx2)
    
    a2y = (Fsy(G, mm, ms, ry) + fy(v2y[i]))/ms
    vy2 = v2y[i] + a2y*dt
    y2 = r2y[i] + vy2*dt
    r2y.append(y2) # m
    v2y.append(vy2)
    
    a1x = Fmx(G, mm, ms, rx)/mm
    vx1 = v1x[i] + a1x*dt
    x1 = r1x[i] + vx1*dt
    r1x.append(x1) # m
    v1x.append(vx1)
    
    a1y = Fmy(G, mm, ms, ry)/mm
    vy1 = v1y[i] + a1y*dt
    y1 = r1y[i] + vy1*dt
    r1y.append(y1) # m
    v1y.append(vy1)

    i = i + 1
```

The last thing to do is to plot the trajectory of beagle 2. and Mars. this is a plot of the center of mass in both plots.
